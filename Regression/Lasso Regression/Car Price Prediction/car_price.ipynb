```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('CarPrice_Assignment.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
int_=data.dtypes=='int64'
int__=list(int_[int_].index)
print("List of INT Values: ")
print(int__)

float_=data.dtypes=='float64'
float__=list(float_[float_].index)
print("List of FLOAT Values: ")
print(float__)

cat_=data.dtypes=='object'
cat__=list(cat_[cat_].index)
print("List of Categorical Values: ")
print(cat__)

```

```python
num_list=int__+float__
plt.figure(figsize=(6,6))
sns.heatmap(data[num_list].corr())
plt.show()
```

```python
# List of Number of Unique Values

for i in int__:
    print("INT- ", i, "- ", data[i].nunique())

for j in float__:
    print("FLOAT- ", j, "- ", data[j].nunique())

for k in cat__:
    print("CAT- ", k, "- ", data[k].nunique())
```

```python
data.drop(columns='car_ID', inplace=True)
```

```python
encoded_data=pd.get_dummies(data, columns=cat__, drop_first=True)
```

```python
encoded_data.columns
```

```python
X=encoded_data.drop(columns='price')

Y= encoded_data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

model=Lasso(alpha=1.0, max_iter=3433)


model.fit(X_train, Y_train)




```

```python
predictions=model.predict(X_test)
```

```python
predictions
```

```python
Y_test=(Y_test>=0.5).astype(int)

predictions_binary= (predictions>=0.5).astype(int)
```

```python
print("Classification Report: ", classification_report(Y_test, predictions_binary))
```

```python
print("Accuracy Score: ", accuracy_score(Y_test, predictions_binary))
```

```python
print("Confusion Matrix: \n", confusion_matrix(Y_test, predictions_binary))
```
