```python
import pandas as pd
dataset_file='/content/dev-v2.0.json'
df=pd.read_json(dataset_file)
df.head()
```

```python
import json
with open(dataset_file, 'r') as file:
    data = json.load(file)
```

```python
dataset = [item for topic in data['data'] for item in topic['paragraphs'] ]
```

```python
from sklearn.model_selection import train_test_split
train_data, eval_data=train_test_split(dataset, test_size=0.2, random_state=42)
```

```python
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
model_args = QuestionAnsweringArgs(
    output_dir="./roberta_squad2",
    num_train_epochs=2,
    train_batch_size=8,
    evaluate_during_training=True
)

model=QuestionAnsweringModel(
    "roberta",
    "roberta-base",
    args=model_args,
    use_cuda=True

)
```

```python
model.train_model(train_data, eval_data=eval_data)
```

```python
results, texts = model.eval_model(eval_data)

results, texts
```

```python
to_predict = [
    {
        "context": "Vin is a Mistborn of great power and skill.",
        "qas": [
            {
                "question": "What is Vin's speciality?",
                "id": "0",
            }
        ],
    }
]

model=QuestionAnsweringModel("roberta", "/content/roberta_squad2", use_cuda=True)
custom_res, custom_text=model.predict(to_predict)
custom_res[0]['answer'][0]
```
